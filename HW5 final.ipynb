When we do statistical tests, we're not just interested in the **sample** (the specific data we collected), but in what it tells us about the **population** (the entire group we're studying).

- **Population**: The whole group we want to learn about.
- **Sample**: A smaller group we collect data from.
- **Population parameter**: A value or characteristic (like the true average) that we want to know about the population but typically don't.
- **Sample statistic**: A value calculated from the sample that we use to estimate the population parameter.

The key point is that our tests help us make inferences about the **population parameter** based on the **sample data**. We use the sample statistic to make educated guesses about the whole population—not just the specific group we collected data from. The ultimate goal is to understand the population through our sample analyses.

## Q3
### written answer

When calculating a **p-value**, we imagine a world where the **null hypothesis** is true. The p-value measures how likely our sample data is if that assumption holds.

- **Null Hypothesis**: Assumes no effect or difference.
- **Sampling Distribution**: Under the null hypothesis, it shows all possible outcomes of our test statistic by random chance.

The p-value tells us how far our actual result is from what we'd expect under the null hypothesis:
- **Small p-value**: Suggests the data is unlikely to have occurred by chance, leading us to **question or reject the null hypothesis**.
- **Large p-value**: Suggests the data is consistent with the null, so we don’t have enough evidence to reject it.

So, in the study by Güntürkün (2003), they looked at how couples tilt their heads when kissing and found that **80 out of 124 couples** (about **64.5%**) tilted to the right. The big question here is: Is this just random chance, or do people actually tend to tilt their heads to the right more often?

To figure this out, we’ll simulate a **p-value** using the idea that there’s **no preference** for right or left tilts (the null hypothesis). In other words, we’re assuming it’s a **50/50 chance** which way people tilt their heads, like flipping a coin.

### Problem Breakdown

Here’s how the problem shakes out:

- **Population**: All people who kiss (pretty big group, right?).
- **Sample**: The 124 couples in this study.
- **Population parameter**: The true percentage of people who tilt their heads to the right when kissing.
- **Sample statistic**: In this case, it’s the **64.5%** (80 out of 124) couples who tilted to the right.

### Null Hypothesis (H₀)

The **null hypothesis** says there’s no preference between tilting left or right—it’s a 50/50 coin flip. If this is true, the data we observed (80 right tilts) would just be the result of randomness.

### Simulating the P-Value

We’re going to use a "coin flip" model to simulate whether people tilt their heads right or left by chance. Here's how we do it:

1. **Imagine the Null Hypothesis World**: Think of each couple as flipping a fair coin: heads (H) = right tilt, tails (T) = left tilt.
   
2. **Simulate**: We’ll simulate this "coin flipping" process for **124 couples** and do this like **10,000 times** to see what kind of results we’d expect if it really were random.

3. **Compare the Actual Data**: Now, let’s compare our real result—**80 right tilts**—to what we see in the simulation. How often do we get 80 or more right tilts just by chance?

4. **P-Value**: The p-value is the proportion of times we get **80 or more** right tilts in our 10,000 simulations. If this number is super small, it means it’s unlikely the data we observed is just due to random chance.

### Simulation Results

After running the simulation, the p-value turns out to be:

\[
p = 0.0012
\]

### What Does That Mean?

Basically, the p-value tells us how rare our result is under the assumption that people have no head tilt preference (50/50 chance). Based on the table below:

| **p-value**        | **Evidence**                                      |
|--------------------|--------------------------------------------------|
| p > 0.10           | No evidence against the null hypothesis           |
| 0.05 < p ≤ 0.10    | Weak evidence against the null hypothesis         |
| 0.01 < p ≤ 0.05    | Moderate evidence against the null hypothesis     |
| 0.001 < p ≤ 0.01   | Strong evidence against the null hypothesis       |
| p ≤ 0.001          | Very strong evidence against the null hypothesis  |

Our p-value is **0.0012**, which gives us **strong evidence** against the null hypothesis (but not quite "very strong").

### Conclusion

What this means is that there’s **strong evidence** that humans don’t just tilt their heads randomly when kissing. The chance of seeing **80 or more** couples tilt their heads to the right if it were truly random is pretty slim. So, we’d **reject the null hypothesis** and say that people **probably do** have a tendency to tilt their heads to the right when they kiss.

#### Relationship to Fisher's Original Experiment

- **Fisher's Experiment**: 8 cups, 4 with milk first, 4 with tea first. Bristol correctly identified all cups.
- **STA130 Experiment**: 80 students, 49 identified correctly.
  
Both experiments aim to test the ability of individuals to discern the difference between the order of tea and milk, with the **null hypothesis** being that the identifications are purely due to **random guessing**.

### Null and Alternative Hypotheses

#### Formal Null Hypothesis (H₀)
- \( H_0 \): Students are randomly guessing, with a 50% probability of correctly identifying whether tea or milk was poured first.
  
#### Formal Alternative Hypothesis (H₁)
- \( H_1 \): Students are not randomly guessing, and there is a preference or ability to correctly identify whether tea or milk was poured first.

#### Informal Hypothesis Explanation
- **Null Hypothesis (H₀)**: The students are just guessing randomly, so we expect about 50% of them to get the correct answer.
- **Alternative Hypothesis (H₁)**: The students have some ability to detect the difference, so more than 50% of them will correctly identify the order.

### Quantitative Analysis

We will use a **hypothesis test** to check if the observed result of 49 out of 80 correct answers could occur by random guessing.

#### Methodology

We assume that if students are guessing, the number of correct identifications follows a **binomial distribution** with:
- \( n = 80 \) (sample size),
- \( p = 0.5 \) (probability of guessing correctly).

We will simulate this using Python and compare the observed result to the distribution of correct guesses under the null hypothesis.


````python
import numpy as np
from scipy import stats

np.random.seed(42)

# Parameters
n_students = 80
observed_correct = 49
p_guess = 0.5

simulated_data = np.random.binomial(n_students, p_guess, 10000)

p_value = np.mean(simulated_data >= observed_correct)

print(f"P-value: {p_value}")
````

### Supporting Visualization 

````python
import matplotlib.pyplot as plt

plt.hist(simulated_data, bins=range(min(simulated_data), max(simulated_data)+1), alpha=0.7, edgecolor='black')
plt.axvline(observed_correct, color='red', linestyle='dashed', linewidth=2, label=f'Observed Correct = {observed_correct}')
plt.title('Distribution of Correct Guesses Under Null Hypothesis')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

#### Findings and Discussion

The p-value from the simulation tells us the likelihood of getting **49 or more correct identifications** purely by random guessing.

- If the p-value is **low** (typically \( p < 0.05 \)), it suggests that the observed result is **unlikely under the null hypothesis**, providing evidence that the students might be able to detect the difference between tea and milk pouring order.
- If the p-value is **high**, it suggests that the result could easily occur by chance, meaning there's **no strong evidence** that the students are doing better than random guessing.

#### Conclusion Regarding the Null Hypothesis

- If the p-value is **small**, we would **reject the null hypothesis** and conclude that there’s evidence the students can detect the difference between tea and milk pouring order.
- If the p-value is **large**, we would **fail to reject the null hypothesis**, meaning the students' performance is consistent with random guessing.


### Conclusion
Based on the p-value from the hypothesis test, we can make a conclusion about whether or not STA130 students have a real ability to detect the difference between the pouring order of tea and milk. This experiment, much like Fisher's original experiment, helps us understand if observed outcomes are due to chance or reflect genuine ability.
